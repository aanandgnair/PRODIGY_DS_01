```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "/content/PI TASK1.csv"  # Replace with the path to your dataset
data = pd.read_csv(file_path, skiprows=4)  # Skip metadata rows
```

```python
data_cleaned = data.drop(columns=["Unnamed: 68"], errors="ignore")

# Display basic info
print(data_cleaned.info())
print(data_cleaned.head())
```

```python
population_data = data_cleaned[data_cleaned["Indicator Name"] == "Population, total"]
```

```python
plt.figure(figsize=(12, 6))
top_countries = population_data.nlargest(10, "2020")
plt.bar(top_countries["Country Name"], top_countries["2020"])
plt.title("Top 10 Countries by Population in 2020")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(population_data["2020"].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Population in 2020")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
country_name = "India"  # Replace with the country you want to analyze
country_data = population_data[population_data["Country Name"] == country_name]
years = [str(year) for year in range(1960, 2024)]
plt.figure(figsize=(14, 7))
plt.plot(years, country_data[years].iloc[0], marker='o')
plt.title(f"Population Trend of {country_name} (1960-2023)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
